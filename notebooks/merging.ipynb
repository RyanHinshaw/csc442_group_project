```python
import pandas as pd
import numpy as np
```

```python
best_seasons = pd.read_csv("best_seasons.csv")
best_seasons = best_seasons.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
best_seasons
```

```python
draft_combine = pd.read_csv('draft_combine_data.csv')
draft_combine = draft_combine.drop('Unnamed: 0', axis=1)
draft_combine
```

```python
merged_df = pd.merge(best_seasons, draft_combine, left_on='player_name', right_on='PLAYER')
merged_df = merged_df.drop(['team_abbreviation', 'college', 'country', 'HANDL', 'HANDW', 'PLAYER', 'YEAR', 'HGT', 'WGT'], axis=1)
merged_df
```
