```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/RyanHinshaw/csc442_group_project/"
 + "refs/heads/main/datasets/merged_player_data.csv")
df = df.sort_values(by='pts', ascending=False).reset_index(drop=True)
df['draft_number'] = pd.to_numeric(df['draft_number'], errors='coerce')
df['lottery_pick'] = df['draft_number'] <= 14
df['POS'] = df['POS'].str.split('-').str[0]
query_df = df[['player_name', 'lottery_pick', 'POS', 'STNDVERT', 'LPVERT', 'LANE', 'SPRINT']]
query_df
```

```python
sns.scatterplot(x='SPRINT', y='LANE', hue='POS', style='lottery_pick',
                markers={True: 'o', False: 'X'}, data=query_df, palette='Set1', s=100)
# Labels and title
plt.xlabel('Lane Drill (seconds)')
plt.ylabel('Sprint Drill (seconds)')
plt.title('Combine Performance: Drill Speeds by Position (Lottery Picks vs Non-Lottery)')
# Show the plot
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='STNDVERT', y='LPVERT', hue='POS', style='lottery_pick',
                markers={True: 'o', False: 'X'}, data=query_df, palette='Set1', s=100)
# Labels and title
plt.xlabel('Standing Vertical Jump (inches)')
plt.ylabel('Leaping Vertical Jump (inches)')
plt.title('Combine Performance: Jumping Drill Heights by Position (Lottery Picks vs Non-Lottery)')
# Show the plot
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
