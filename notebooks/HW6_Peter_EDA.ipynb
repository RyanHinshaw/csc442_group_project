```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('../datasets/merged_player_data.csv')

# Get the player height, weight, position, rebounds, and assists, and drop all rows with a null value
query_df = df[['player_height', 'player_weight', 'POS', 'reb', 'ast']].dropna()

# Display the first few rows
query_df.head()
```

```python
# Change the POS column to remove everything with a '-' and after
query_df['POS'] = query_df['POS'].str.split('-').str[0]
```

```python
# Plot the distribution of rebounds vs assists by position
plt.figure(figsize=(12, 6))
sns.scatterplot(data=query_df, x='reb', y='ast', hue='POS', alpha=0.7)
plt.title('Scatter plot of rebounds vs. assists colored by position')
plt.xlabel('Rebounds')
plt.ylabel('Assists')
plt.legend(title='Position')
plt.grid()
# Restrict this to -1,10 to remove outliers from the view
plt.xlim(-1, 10)
plt.ylim(-1, 10)
plt.show()

```

```python
# Plot the distribution of height vs weight by position
plt.figure(figsize=(12, 7))
sns.scatterplot(data=query_df, x='player_weight', y='player_height', hue='POS', alpha=0.7)
plt.title('Scatter plot of player height vs. weight colored by position')
plt.xlabel('Player Height')
plt.ylabel('Player Weight')
plt.legend(title='Position')
plt.grid()
plt.show()
```
